```python
import numpy as np
```

```python
data["percent_grade_bucket"] = np.ceil(data["percent_grade"]/0.2)
```

```python
data["percent_grade_bucket"].where(data["percent_grade_bucket"] < 5, 5.0, inplace = True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
```

```python
c = data.copy()
```

```python
c.head()
```

```python
values = {'week':0, 'all_clicks':0, 'video_plays':0, 'grade':0, 'grade_diff':0, 'time_diff':0, 'percent_grade':0, 'seq_next_count':0, 'seq_goto_count':0, 'play_video_count':0, 'pause_video_count':0, 'closed_captions_show_count':0, 'problem_save_count':0, 'seek_video_count':0, 'link_clicked_count':0, }
c = c.fillna(value = values)
```

```python
for train_index, test_index in split.split(data, data['percent_grade_bucket']):
    strat_train_set = c.loc[train_index]
    strat_test_set = c.loc[test_index]
```

```python
strat_train_set
```

```python
data_training_label = strat_train_set['percent_grade']
data_training_attributes = strat_train_set.copy()
data_training_attributes = data_training_attributes.drop(['percent_grade','percent_grade_bucket', 'course_id', 'passed', 'time_diff'], axis=1)
```

```python
data_training_attributes
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(data_training_attributes, data_training_label)
```

```python
data_testing_label = strat_test_set['percent_grade']
data_testing_attributes = strat_test_set.copy()
data_testing_attributes = data_testing_attributes.drop(['percent_grade','percent_grade_bucket', 'course_id', 'passed', 'time_diff'], axis=1)
```

```python
from sklearn.metrics import mean_squared_error
data_predictions = lin_reg.predict(data_training_attributes)
lin_mse = mean_squared_error(data_predictions, data_training_label)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(data_training_attributes, data_training_label)
```

```python
data_predictions = tree_reg.predict(data_training_attributes)
tree_mse = mean_squared_error(data_training_label, data_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_mse
```
